# <a id='toc1_'></a>[用Python进行数学建模](#toc0_)
---

```python
import sympy as sp

A = sp.Matrix([[1], [2], [3]])
B = sp.Matrix([[4], [5], [6]])

print(f'A的模为：{A.norm()}')
print(f'A的模的浮点数为：{A.norm().evalf()}')
print(f'A和B的点乘积为：{A.dot(B)}')
print(f'A和B的叉乘积为：{A.cross(B)}')
```

```python
import numpy as np
import sympy as sp

A = sp.Matrix(np.arange(1,17).reshape(4, 4)); print(A)
B = sp.eye(4); print(B)
print(f'A的行列式：{sp.det(A)}')
print(f'A的rank：{A.rank()}')
print(f'A的转动置：{A.transpose()}')
```

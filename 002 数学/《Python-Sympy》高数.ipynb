```python
# x, y, z, t = sympy.symbols('x,y,z,t') # 无限制符号
# k, m, n = sympy.symbols('k,m,n', integer=True) # k,m,n 为整数符号
# f, g, h = sympy.symbols('f,g,h', cls=Function) # f,g,h 为函数符号

# 符号
x = sympy.Symbol('x')
y = sympy.Symbol('y')
z = sympy.Symbol('z')
# 等同于
x, y, z = sympy.symbols('x, y, z')
```

欧拉恒等式，其中e是自然
常数，i是虚数单位， 是圆周率。此公式被誉
为数学中最奇妙的公式，它将5个基本数学常数
用加法、乘法和幂运算联系起来。
从SymPy库载入的符号中，E表示自然常
数，I表示虚数单位，pi表示圆周率，因此上面
的公式可以直接如下计算：


```python
f = (sympy.E)**(sympy.I*sympy.pi) + 1 # sympy.I 为虚部
```

## <a id='toc3_2_'></a>[simplify 合并同类项](#toc0_)

```python
f = (2/3)*x**2 + (4/3)*x**2 + x + 5*x + 10 
print(f'公式1：{f}') # 自动合并同类项了
print(f'公式2：{f.simplify()}') # 利用sympy。simplify进行简化（合并同类项）
f.simplify()
sympy.plot(f) # 还可以绘图，更加直观
```

## <a id='toc3_3_'></a>[expand 展开](#toc0_)

```python
f = (x + y)**3
print(f'公式1：{f}') # 原公式
print(f'公式2：{f.expand()}') # 展开后的公式
f.expand()
```

## <a id='toc3_4_'></a>[plot 绘图](#toc0_)

```python
# 支持绘图更加直观
f1=x+sympy.sin(x) 
f2=x
f3=x-sympy.sin(x)
f4=-x+sympy.sin(x) 
f5=-x
f6=-x-sympy.sin(x)
sympy.plot(f1, f2, f3, f4, f5, f6)

sympy.plot((f1,(x,-5,5)), (f2,(x,-12,12)), title='对不同表达式分别指定绘图区间范围')
```

## <a id='toc3_5_'></a>[summation 求和](#toc0_)

### <a id='toc3_5_1_'></a>[累加求和∑](#toc0_)
  

```python
n = sympy.Symbol('n')
f = (1/2)**n
sympy.summation(f, (n, 0, sympy.oo)) # 累加，直接给结果
```

```python
f = (1/sympy.log(n, 10)) + 3
sympy.summation(f, (n, 0, sympy.oo)) # 计算不出来就直接给累加公式
```

### <a id='toc3_5_2_'></a>[双重累加求和∑∑](#toc0_)

```python
m = sympy.Symbol('m')
f = 2*n + 1
sympy.summation(f, (n, 0, m), (m, 0, 100)) # 直接给结果
```

## <a id='toc3_6_'></a>[solve 方程自动求解](#toc0_)

```python
# 解三元一次方程
f1 = 2*x - y + z - 10
f2 = 3*x + 2*y - z - 16
f3 = x + 6*y - z - 28

print(f'方程解为：{sympy.solve([f1, f2, f3])}')
sympy.solve([f1, f2, f3])
```

## <a id='toc3_7_'></a>[limit 求极限](#toc0_)

```python
f = sympy.sin(x)/x
print(f'极限为：{sympy.limit(f, x, 0)}')
sympy.limit(f, x, 0)
```

## <a id='toc3_8_'></a>[diff 求导](#toc0_)

```python
# diff(你的函数，自变量，求导的次数）
sympy.diff(sympy.sin(2*x), x)
```

# 概率论与梳理统计
---  

```python
import torch 
import matplotlib.pyplot as plt 


def bernoulli(n, p):
    """Bernoulli distribution"""
    return torch.bernoulli(torch.full((n,), p))   

def poisson(n, lam):
    """Poisson distribution"""
    return torch.poisson(torch.full((n,), lam))

bernoulli(10, 0.5)
poisson(10, 5.0)
```

```python
nrows, ncols = 2, 2
plt.subplot(nrows, ncols, 1)
plt.plot(bernoulli(1000, 0.5).numpy(), 'o')
plt.title('Bernoulli Distribution')

plt.subplot(nrows, ncols, 2)
plt.plot(poisson(1000, 5.0).numpy(), 'o')
```

```python
def U(a, b):
    """Uniform distribution"""
    return a + (b - a) * torch.rand(1)


U(0, 1)
```

```python
nrows, ncols = 2, 2
plt.subplot(nrows, ncols, 1)
plt.plot(U(0, 1).numpy(), 'o')
plt.title('Uniform Distribution')
```

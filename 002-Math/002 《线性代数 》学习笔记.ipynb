# 线性代数
---  

- 1. 向量：有向线段-在坐标系中用单位正交向量表示所有其他向量；
- 2. 行和列数相等的矩阵-一维、二维、三维...坐标中的有向线段；
- 3. 行和列数不相等的矩阵-列数表示坐标维度，行数<列数表示降维，反之则升维；
- 4. 矩阵相乘（从右向左）-坐标的线性变换（以相同原点，坐标网格平行且等距。包括坐标系旋转、剪切等等）

# 线性优化

```python
from scipy.optimize import linprog
```

```python
help(linprog)
```

```python
import sympy
x1, x2 = sympy.symbols('x1, x2')
f1, f2, f3, z = sympy.symbols('f1, f2, f3, z', cls=sympy.Function)
f1 = -3*x1 + x2
f2 = x1 + 2*x2
f3 = x2
z = -x1 + 4*x2
sympy.plotting.plot3d(f1, f2, f3, z)
```

```python
c = [-1, 4]
A = [[-3, 1],
     [1, 2]]
b = [6, 4]
bounds=((None,None), (-3,None))
res=linprog(c, A, b, None, None, bounds)
print('目标函数的最小值：', res.fun)
print('最优解：', res.x)
```

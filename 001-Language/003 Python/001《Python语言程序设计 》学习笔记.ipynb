## 1.4. <a id='toc1_4_'></a>[数据类型](#toc0_)

为什么要分不同的数据类型？  
- 计算机中不同的数据类型存储方式不同，各自计算原理不同，为充分利用存储空间和高效计算，遂有多种类型如：字符、数值和布尔等等）。
  - 1. **字符型**（string）
  - 2. **数值型**（整型 int 、浮点型 float ）
  - 3. **布尔型**（True、False）
  - 4. **转换**：str(), int(), float()

````python
# 字符型
s1, s2, s3 = 'You are so beautiful/', "handsome/", '''Yes/'''
print(type(s1), s1)
print(type(s1), s2)
print(type(s1), s3)

# 数值型
a, b, c = 1, 2.1, 3.333
print(type(a), a)
print(type(b), b)
print(type(c), c)

# 布尔型
x, y = True, False
print(type(x), x)
print(type(y), y)
````

### 1.6.2. <a id='toc1_6_2_'></a>[字符串操作](#toc0_)

````python
s1 = 'You are/'
s2 = 'beautiful'

print('无(支持转义字符):', 'How are you \n')

print('u-Unicode(后面字符串以 Unicode 格式 进行编码，支持转义字符):', u'How are you \n')     # u

print('f-Format(格式化输出，支持转移字符):', f'How are you {s1} \n')                        # f

print('r-Remove(去除转义字符，不支持转义字符):', r'How are you \n')                         # r

print('b-Bytes(使后面字符串是bytes 类型，不支持转义字符):', b'How are you \n')               # b
````

````python
## 增
print('-:'*100)                      # * 复制
print('+:', s1 + s2)                 # + 连接字符串
print('*:', s1*3)                   
print('join:', s1.join('Jackie'))    # join 将前字符插入括号中的字符串中
print('join:', '|'.join(s1))
````

````python
## 删
print('-'*100)
print('lstrip', s1.lstrip(), '删除左边空格')    # lstrip
print('rstrip', s1.rstrip(), '删除右边空格')    # rstrip
print('strip', s1.strip(), '删除两边空格')      # strip
````

````python
## 查
print('-'*100)
print('索引', s1[1])
print('切片0:3 ', s1[0:3])
print('切片0:末尾 ',s1[0:])
print('倒序索引 ',s1[-1])
print('切片-左闭右开',s1[:-1])

print('startswith',s1.startswith('Y'))  # startswith
print('endswith',s1.endswith('/'))      # endswith
print('find,返回下标',s1.find('/'))      # find
print('in', 'You' in s1)                # in
print('not in', 'You' not in s1)        # not in
print('==比较', 'You' == s1)            # ==
print('isalnum', s1.isalnum(), '若字符串中含有数字则返回True，否则为False') # isalnum
print('isalpha', s1.isalpha(), '若字符串中含有字母则返回True，否则为False') # isalpha
print('islower', s1.islower())          # islower
print('isupper', s1.isupper())          # isupper
print('isspace', (s1+s2).isspace())     # isspace
print('count', s1.count('Y'))           # count
````

````python
## 改
print('-'*100)
print('lower',s1.lower())                   # lower
print('upper',s1.upper())                   # upper
print('title',s1.title())                   # title
print('capitalize',s1.capitalize())         # capitalize
print('swqpcase',s1.swapcase())             # swapcase
print('split,默认是以tab或空格为分割符',s1.split()) # split
print('replace',s1.replace('Y', 'YYYYYY'), '并不改变原字符串内容，只是一个复制', s1)    # replace
````

### 1.6.3. <a id='toc1_6_3_'></a>[列表操作](#toc0_)

````python
l = [1, 2, 3]

## 增 insert, append, extend
print('-'*100)
print('+', l + [11, 22, 33 ,44, 55])
print('*', l * 5)
l.insert(0, 0); print(l)                    # 第0号位置插入0
l.append(5); print(l)                       # 列表末尾追加元素5
l.append([11, 22, 33]); print(l)            # 将新列表整体作为l的一个元素追加到末尾
l.extend([11, 22, 33]); print(l)            # 将新列表的元素挨个提取出来并以此追加到l列表中

## 删 reomve
print('-'*100)
l.remove(5); print(l)                       # 从列表中删除查找的第一个匹配元素并删除，若未查找到则引发异常

## 查
print('-'*100)
l[1]                                        # 下标索引
l[0:3]                                      # 序列切片，左闭右开
print('切片', l[0:4:1] )                    # 序列切片，左闭右开，步长为1
print('in', 1 in l)
print('not in', 1 not in l)

## 改
print('-'*100)
l[1] = 11; print(l)                         # 直接修改
````

### 1.10.3. <a id='toc1_10_3_'></a>[继承：](#toc0_)

- 父类、子类（派生类）；

- 父类包含于子类中，因为子类继承了所有父类的属性方法，同时又可以更改；

- super().init，自动执行父类中def init(self)这个方法；

- 更改子类只需要将父类同名方法重写一遍便是覆盖。

- 关于继承更多知识参阅：https://zhuanlan.zhihu.com/p/285244974

- 单继承，多继承（按顺序取值）；父类名字或super()替代父类名称；

- super()替代父类的名字，避免更改父类名称后，继承该父类的子类大量的改名工作量。

````python
class Animal():
    '''creating animal class'''
    def __init__(self):
        self.special = 'Aniaml'
    def eat(self):
        print('eat')
    def drink(self):
        print('drink')
    def sleep(self):
        print('sleep')

class Dog(Animal):
    '''creating dog class from animal'''
    def __init__(self, name):
        self.name = name
        self.special = 'Dog'
    def bark(self):
        print('狗汪汪叫')

class XiaoTianQuan(Dog):
    def __init__(self):
        self.special = '哮天犬'
    def bark(self):
        print('像神一样的狂吠')
        super().bark()                      # 可以保留父类的bark（）方法
        print('hahahahahhah')

class Cat(Animal):
    '''creating cat class from animall'''
    def __init__(self, name):
        self.name = name
        self.special = 'Cat'
    def cry(self):
        print('呜呜。。。。。。')

if __name__ == '__main__':
    wang_cai = Dog('wang cai')             # 类的实例化成对象
    wang_cai.eat()
    wang_cai.bark()
    mao = Cat('mao')                       # 类的实例化成对象
    mao.cry()
    xtq = XiaoTianQuan()                   # 类的实例化成对象
    xtq.bark()
````

### 1.10.4. <a id='toc1_10_4_'></a>[多态:](#toc0_)

- 多态是指一类事物有多种形态，比如动物类，可以有猫，狗，猪等等。（一个抽象类有多个子类，因而**多态的概念依赖于继承**）

````python
class Animal():
	def __init__(self, name):
		self.name = name
	def run(self):
		print("I am runing!")
	def getName(self):
		print('My name is {}'.format(self.name))

class Cat(Animal):
	def Say(self):
		print(' I am a cat')
class Dog(Animal):
	def Say(self):
		print('I am a dog')
class Person(Animal):
	def Say(self):
		print(' I am a person')
````

- **多态性** 是指具有不同功能的函数可以使用相同的函数名，这样就可以用一个函数名调用不同内容的函数。

- 在面向对象方法中一般是这样表述多态性：向不同的对象发送同一条消息，不同的对象在接收时会产生不同的行为（即方法）。也就是说，每个对象可以用自己的方式去响应共同的消息。所谓消息，就是调用函数，不同的行为就是指不同的实现，即执行不同的函数。

````python
class Animal():
	def __init__(self, name):
		self.name = name
	def run(self):
		print("I am runing!")
	def getName(self):
		print('My name is {}'.format(self.name))
	def Say(self):
		pass

class Cat(Animal):
	def Say(self):
		print(' I am a cat')
class Dog(Animal):
	def Say(self):
		print('I am a dog')
class Person(Animal):
	def Say(self):
		print(' I am a person')

cat = Cat('mao')
dog = Dog('gou')
human = Person('human')

def function(obj):
	obj.Say()
	return None
function(cat)
function(dog)
function(human)
````

- 综上可以说，多态性是 一个接口,多种实现

- 通过方法重写和方法重载，我们可以在不同的子类中实现相同的方法，但具体行为可能不同


多态性的好处:  
- 增加了程序的灵活性,以不变应万变，不论对象千变万化，使用者都是同一种形式去调用，如func(obj)
- 增加了程序额可扩展性,通过继承animal类创建了一个新的类，使用者无需更改自己的代码，还是用func(obj)去调用

##### 1.14.2.2.1. <a id='toc1_14_2_2_1_'></a>[迭代器类](#toc0_)
- 类里面包括__iter__() 和 __next__() 

````python
# "__iter__()" 方法返回对象本身，它是for 语句使用迭代器的要求。
# "__next__()" 方法用于返回容器中下一个元素或数据。当容器中的数据用尽时，应该引发Stoplteration 异常。
````

````python
class Frange():
    def __init__(self, start, stop, step) -> None:
        self.__start = start
        self.__stop = stop 
        self.__step = step 
        
    def __iter__(self):
        return self 
    
    def __next__(self):                 # 会自动迭代 (循环)
        if self.__start < self.__stop:
            self.__start += self.__step 
            return self.__start
        else:
            raise StopIteration

test = Frange(0, 2, 0.5)
print(next(test)); print('='*10)
print(next(test)); print('='*10)
print(next(test)); print('='*10)
print(next(test)); print('='*10)
# print(next(test)); print('='*10)
````

````python
def frange(start, stop, step)->float:
    x = start
    while x < stop: # 没有__next__() 只能用while
        yield  x
        x += step

# for
for i in frange(1, 3, 0.5):
    print(i)

# next()
print('='*10)
g = frange(1, 3, 0.5)   # 必须，先实例化
print(next(g))
print(next(g))
print(next(g))
print(next(g))

print('='*10)
print(next(frange(1, 3, 0.5)))
print(next(frange(1, 3, 0.5)))
print(next(frange(1, 3, 0.5)))
print(next(frange(1, 3, 0.5)))
````

#### 1.14.3.2. <a id='toc1_14_3_2_'></a>[方式二：使用yield函数](#toc0_)
- 迭代器函数 = 生成器函数

* 生成器函数：含有yield的函数
  * yield是一个关键字，类似于return关键字，但是用法不同

  * 生成器函数：使用了yield的函数被称为生成器函数。生成器函数返回一个可迭代的生成器对象

  * 生成器对象：生成器对象可以使用for循环或者调用next()方法来提取结果
  
  * 暂停和恢复：在调用生成器函数的过程中，每次遇到yield时函数会暂停并保存当前所有的运行信息，返回yield的值。在下一次执行next()方法时从当前位置继续运行，直到生成器被全部遍历完

````python
def frange(start, stop, step)->float:
    x = start
    while x < stop: # 没有__next__() 只能用while
        yield  x
        x += step

# next()
g = frange(1, 3, 0.5)   # 必须，先实例化
print(next(g))
print(next(g))
print(next(g))
print(next(g))

print('='*10)
print(next(frange(1, 3, 0.5)))
print(next(frange(1, 3, 0.5)))
print(next(frange(1, 3, 0.5)))
print(next(frange(1, 3, 0.5)))
````

## 1.16. <a id='toc1_16_'></a>[闭包（closure）](#toc0_)

- 在一个内部函数中，对外部作用域的变量进行引用，并且一般外部函数的返回值为内部函数，那么内部函数就被认为是闭包。

- **在函数中定义函数**。

````python
def out(a, b): 						# 外部函数
	def inner(x): 					# 内部函数
		return a * x + b
	return inner

line = out(3, 5)  					# 调用，给定斜率和截距

print(f'line的数据类型：{type(line)}, 其实就是inner')

print(f'line(inner)在引用了out的变量后，再传自己要的参数并计算结果为： {line(30)}')		# 调用，输入x值并计算
````

- 遵循变量作用域规则

````python
def out1():
	x = 1
	print(f'out x: {x}')
	def inner1():
		x = 2
		print(f'inner x: {x}')
	inner1()
	print(f'out x after call inner1: {x}')
	return inner1
out1() 										# 内部函数不可以修改外部函数的变量内容，变量作用域
````

- 可以嵌套超过两层

````python
# 可以嵌套多层函数，原理也简单-就是函数
def first():
    a = 1
    print(f'first a: {a}')
    def second():
        b = 2
        print(f'second b: {b}')
        def third():
            c = 3
            print(f'third c: {c}')
            print(f'最内层函数照样能调用所有外层变量，a:{a}, b: {b}, c: {c}')
            return None
        return third
    return second

first()                                     # 只能执行first
print('-'*100)
first()()                                   # 执行first后再执行second
print('-'*100)
first()()()                                 # 执行first，再执行second，再执行third
````

## 1.17. <a id='toc1_17_'></a>[装饰器（Decorator）](#toc0_)

- “don’t repeat yourself”，当你的程序中存在高度重复(或者是通过剪切复制)的代码时，都应该想想是否有更好的解决方案；

- 在Python当中，通常都可以通过 **元编程** 来解决这类问题： **装饰器** 的应用-元编程；

- 简而言之，元编程就是关于创建操作源代码(比如修改、生成或包装原来的代码)的函数和类。 主要技术是使用装饰器、类装饰器和元类。不过还有一些其他技术， 包括签名对象、使用 exec() 执行代码以及对内部函数和类的反射技术等。

- 什么是装饰器呢，装饰器是闭包的一种应用。

- 一个装饰器就是一个函数，它接受一个函数作为参数并返回一个新的函数。

- 例如：你想在函数上添加一个包装器，增加额外的操作处理(比如日志、计时等)。

````python
import time

def timer(func): 									# 装饰函数定义
	def wrapper(*args, **kwargs):
		t1 = time.time()
		func()
		t2 = time.time()
		print('执行所用时间为：{}'.format(t2 - t1))
	return wrapper

@timer 												# 语法糖
def have_a_sleep(): 								# 被装饰函数定义
	time.sleep(5)

have_a_sleep() 										# 调用

# 如下同样可以实现调用，只是并不简洁
dy = timer(have_a_sleep)
dy()
````

- 讨论：
  - 一个装饰器就是一个函数，它接受一个函数作为参数并返回一个新的函数。 

- 问题：你写了一个装饰器作用在某个函数上，但是这个函数的重要的元信息比如名字、文档字符串、注解和参数签名都丢失了。
- 解决：任何时候你定义装饰器的时候，都应该使用 functools 库中的 @wraps 装饰器来注解底层包装函数。例如：

````python
import time
from functools import wraps

def timethis(func):
    '''Decorator that reports the execution time.'''
    @wraps(func)                                # ------------------------------------------ # 
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(func.__name__, end-start)
        return result
    return wrapper

@timethis
def countdown(n):
    '''Counts down descriptiong or annotation'''
    while n > 0:
        n -= 1

countdown(100000)
print('countdown.__name__: ', countdown.__name__)
print('countdown.__doc__: ', countdown.__doc__)
print('countdown.__annotations__: ', countdown.__annotations__)
````

- 问题：你想定义一个可以接受参数的装饰器；
- 解决：我们用一个例子详细阐述下接受参数的处理过程。 假设你想写一个装饰器，给函数添加日志功能，同时允许用户指定日志的级别和其他的选项。 下面是这个装饰器的定义和使用示例：

````python
from functools import wraps
import logging

def logged(level, name=None, message=None):
    """
    Add logging to a function. level is the logging
    level, name is the logger name, and message is the
    log message. If name and message aren't specified,
    they default to the function's module and name.
    """
    def decorate(func):
        logname = name if name else func.__module__
        log = logging.getLogger(logname)
        logmsg = message if message else func.__name__

        @wraps(func)
        def wrapper(*args, **kwargs):
            log.log(level, logmsg)
            return func(*args, **kwargs)
        return wrapper
    return decorate

# Example use
@logged(logging.DEBUG)
def add(x, y):
    return x + y

@logged(logging.CRITICAL, 'example')
def spam():
    print('Spam!')
````

- 讨论：
  - 初看起来，这种实现看上去很复杂，但是核心思想很简单。 最外层的函数 logged() 接受参数并将它们作用在内部的装饰器函数上面。 内层的函数 decorate() 接受一个函数作为参数，然后在函数上面放置一个包装器。 **这里的关键点是包装器是可以使用传递给 logged() 的参数的。**
  - 定义一个接受参数的包装器看上去比较复杂主要是因为底层的调用序列。特别的，如果你有下面这个代码：

````python
@decorator(x, y, z)
def func(a, b):
    pass

# 装饰器处理过程跟下面的调用是等效的;
def func(a, b):
    pass
func = decorator(x, y, z)(func) # decorator(x, y, z) 的返回结果必须是一个可调用对象，它接受一个函数作为参数并包装它
````

- 总结：
  - 不带参数的装饰器：用两层嵌套即可；
  - 记得带上@wraps()保护原函数相关信息；
  - 带参数的装饰器：用三层嵌套即可，最外层接受收参数并作用于所有下层函数。@logged(x, y, z) 对应于 def logged(x, y, z):

### 1.20.3. <a id='toc1_20_3_'></a>[多进程编程-Process](#toc0_)
#### 1.20.3.1. <a id='toc1_20_3_1_'></a>[concurrent.futures.ProcessPoolExecutor()的进程池](#toc0_)

````python
import math
import time
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor

def isPrime(n:int)->None:
    '''阶乘'''
    mulTotal = 1
    for i in range(1, n + 1):
        mulTotal *- i
    # print(mulTotal)

def com(PRIMES:list)->None:
    '''设置PRIMES大小（计算量大小）比较单线程、多线程和多进程各自效率'''

    print('PRIMES: ', PRIMES[0])

    # 单线程
    start = time.time()
    for i in PRIMES:
        isPrime(i)
    stop = time.time()
    print('单线程执行时间：', stop - start, 'seconds')

    # 多线程
    start = time.time()
    with ThreadPoolExecutor() as pool:
        pool.map(isPrime, PRIMES)                       # PRIMES是列表
    stop = time.time()
    print('多线程执行时间：', stop - start, 'seconds')

    # 多进程
    start = time.time()
    with ProcessPoolExecutor() as pool:
        pool.map(isPrime, PRIMES)                       # PRIMES是列表
    stop = time.time()
    print('多进程执行时间：', stop - start, 'seconds')

# 在计算任务不大的时候并发编程可能更耗时,尤其是多线程对CPU密集型任务效率提升不高
# 计算任务和大时候，并发编程可以提高执行效率
PRIMES = [100] * 100; com(PRIMES)
PRIMES = [10000] * 100; com(PRIMES)
PRIMES = [10000000] * 100; com(PRIMES)
````

##### 1.22.2.2.2. <a id='toc1_22_2_2_2_'></a>[subprocess.check_output()-执行外部命令并获取它的输出](#toc0_)

````python
#正常执行

import subprocess

# 执行一个指定的命令并将执行结果以一个字节字符串的形式返回。
## 默认情况下，check_output() 仅仅返回输入到标准输出的值。
out_bytes = subprocess.check_output('ipconfig -all') 
## 如果你需要同时收集标准输出和错误输出，使用 stderr 参数：                           
out_bytes = subprocess.check_output('ipconfig -all', stderr=subprocess.STDOUT)  
print(f'原始的out_bytes: {out_bytes}')

# 如果你需要文本形式返回，加一个解码步骤即可。例如：
## 'gbk', 'utf-8'
print('='*10)
print(f"解码后的out_text: {out_bytes.decode('gbk')}")
````

````python
# 执行失败，程序会崩溃

import subprocess

cmd = 'ipconfig --error'                         

out_bytes = subprocess.check_output(cmd, stderr=subprocess.STDOUT)
print(f'原始的out_bytes: {out_bytes}')

out_text = out_bytes.decode('gbk')                  # 'utf-8', 'gbk'
print(f'解码后的内容： {out_text}')
````

###### 1.22.2.2.2.1. <a id='toc1_22_2_2_2_1_'></a>[推荐：如果被执行的命令以非零码返回，就会抛出异常。 下面的例子捕获到错误并获取返回码：](#toc0_)

````python
import subprocess

try:
    # cmd = 'ipconfig -all'                          # 正常执行
    cmd = 'ipconfig --error'          # 引起异常

    out_bytes = subprocess.check_output(cmd, stderr=subprocess.STDOUT)
    print('程序正常执行，返回值为：0')

    out_text = out_bytes.decode('gbk')                  # 'utf-8', 'gbk'
    print(f'Decode contents: {out_text}')

except subprocess.CalledProcessError as e:
    print('执行失败。')
    print(f'返回值为：{e.returncode}')
    print(f"e的输出值为：{e.output.decode('gbk')}")
````

###### 1.22.2.2.2.2. <a id='toc1_22_2_2_2_2_'></a>[subprocess.check_output()](#toc0_)
- 使用 check_output() 函数是执行外部命令并获取其返回值的最简单方式;  

- 但是，如果你需要对子进程做更复杂的交互，比如给它发送输入，你得采用另外一种方法。 这时候可直接使用 subprocess.Popen 类。例如：

````python
import subprocess

# Some text to send
text = b'''
hello world
this is a test
goodbye
'''

# Launch a command with pipes
p = subprocess.Popen('echo',
          stdout = subprocess.PIPE,
          stdin = subprocess.PIPE)

# Send the data and get the output
stdout, stderr = p.communicate(text)

# To interpret as text, decode
out = stdout.decode('gbk')
err = stderr.decode('gbk')
````

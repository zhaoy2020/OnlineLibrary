- https://blog.csdn.net/woodcorpse/article/details/106552561
- 从研究对象上来看，生态学的研究水平，主要包括个体、种群、群落和生态系统四个水平；根据《基础生态学》的定义：
  - 种群(population)是指，同一时期内一定空间中同种生物个体(individual)的集合；
  - 群落(community)是指，相同时间聚集在同一地段的各物种种群的集合；
  - 生态系统(ecosystem)是指在一定空间中共同栖居的所有生物与其环境之间由于不断进行物质循环和能量流动过程而形成的统一整体。

- 什么是群落测度？
  - 微生物群落的测度（measure）是指对群落矩阵数据的一种度量比较。测度可以用一系列指数（index）或系数（coefficient）来表示。对于单个对象（样品）的测度计算，可以采用α多样性指数来表示，而对于不同对象之间的比较，则可以采用β多样性指数或者距离。对于变量（物种或环境因子）之间的比较，则采用相关性来比较。群落测度的分析结果，可用于后续的排序分析、网络分析、聚类分析、判别分析等。
- α多样性指数简介
 - α多样性指数反应群落内物种数量及其相对丰度，为群落内各物种利用同一生境互相竞争或共生的结果，比较不同样本的α多样性指数可以看出不同样本多样性差异。
 - α多样性指数反应群落内物种数量及其相对丰度，为群落内各物种利用同一生境互相竞争或共生的结果，比较不同样本的α多样性指数可以看出不同样本多样性差异。
   - ①丰富度指数(communityrichness)反映的是群落内物种的丰富程度，常见的有sobs、chao、ace、jack、bootstrap等。sobs即根据测序结果在一定抽样水平实际观察到的物种数目（observedOTUs），chao、ace则分别用chao1和ace算法来估算一定抽样水平物种的总数目。
   - ②均匀度指数(communityevenness)反映的是群落内物种的分配状况，也即各物种相对丰度的均匀程度，在同一群落内与丰富度指数呈正相关，常见的有simpsoneven、shannoneven、heip、smithwilson等。
   - ③物种多样性指数(communitydiversity)反映的是物种丰富度和均匀度的综合状况，常见的有shannon-wienner（也即shannon）、simpson、invsimpson等。shannon指数反映的是物种丰度与均匀度，与这两者均呈正相关；simpson指数为在样本中抽取两条序列属于不同种的概率。
   - 物种多样性指数为丰富度与均匀度的综合考察，其主要参数介绍如下:
   

- α多样性：是指一个群落内物种的个数(species richness, 丰富度)以及每个物种的数量及分布(evenness,均匀度)。
  - 度量α-多样性的指数有很多，常用的如Simpson index, Shannon-Weiner index, Pielou evenness, Simpson evenness等。
  - 随着二代测序的广泛应用，基于序列数据而出现的Chao1 value, OTU number等，也被推广使用。
  - 一定程度上，α-多样性依然是最直接、最直观的群落比较方式。比如，食性单一、爱吃竹子的大熊猫，其肠道微生物的多样性明显低于其他动物，一定程度上说明食物的单调与肠菌的单调是相关的(图3)。
  - 然而需要注意的是，比较α-多样性的前提是，尽量保证取样量是相当的。比如做物种-样品个数(测序量)的稀疏曲线(图4)，曲线基本都达到平台期时，比较才有意义。
    - 反映群落丰富度（Community richness）的指数有：sobs、chao、ace、jack、bootstrap；
    - 反映群落均匀度（Community evenness）的指数有：simpsoneven、shannoneven、heip、smithwilson；
    - 反映群落多样性（Community diversity）的指数有：shannon、simpson、npshannon、bergerparker、invsimpson、qstat；
    - 反映群落覆盖度（Community coverage）的指数有：coverage。
    - 各指数介绍参考：http://www.mothur.org/wiki/Calculators，具体说明如下：
    
    
    
    

- β-多样性:标准定义是，度量在地区尺度上物种组成沿着某个梯度方向从一个群落到另一个群落的变化率。
  - 度量β-多样性的指标也非常多，比如最常用的基于Bray-curtis距离的计算方式，主要考虑物种的数量与丰度。而微生物群落分析常用的weighted以及unweighted UniFrac，则是结合了物种的进化关系。
  - 这个指标说白了是从另一个角度比较群落跟群落的区别，上一个指标是综合的指数，而这个指标则是以矩阵的数据形式，逐一比较 **每个物种** 在两个群落的数量和分布的差异。
  - 与α-多样性类似，β-多样性也是可以比较的(图5)。例如，β-多样性低的群落，说明其波动性小、其各样点在散点图中会距离比较近。

- γ-多样性:以倪晋仁老师组沿着长江取样来举个例子(图6)，图中每个取样点本身的微生物多样性是α-多样性（单个群落水平）；将1～9这九个样点归为一组，则这一组内各样点之间的差异，可以理解为β-多样性（局部或者区域水平）；当我们把所有取样点放在一起，站在一个更高、更广阔的尺度上时，这时候的多样性就是所谓的γ-多样性。实际上，我个人的理解这个指标就是α-多样性在尺度上的一个推广。

- 尺度（参照系）：最后特别想说明的一点是，上述指标的计算都是相对的。即，你自己定义的范围决定了你最后计算出来的值。还是上面这张图，你也可以把1～9全部放在一起，定义为一个群落，计算它的α-多样性，或者把整条长江定义为一个群落，计算它的α-多样性。另外两个指标，以此类推。只要你的假设和解释是合理的，所有的计算都是有意义的。因为生态学的研究里，尺度(图7)是非常重要的，也是非常基础的一个理念，需要时刻谨记的。不同的研究目标、不同的假设或者预期，可能会需要你调整尺度、或者选择合适的尺度，这样也就导致计算的多样性的数值是相对的，而没有绝对的对错。

### 3.2.1. <a id='toc3_2_1_'></a>[两组比较](#toc0_)
- Student t检验（Student's t test），亦称T检验，是用t分布理论来推论差异发生的概率，从而比较两个平均数的差异是否显著，主要用于样本含量较小（例如n<30），总体标准差σ未知的正态分布数据
- Wilcoxon秩和检验（Wilcoxon rank-sum test），也叫曼-惠特尼U检验（Mann–Whitney U test），是两组独立样本非参数检验的一种方法。
- Wilcoxon signed-rank test，是两组配对样本非参数检验的一种方法。

```bash
ANCOM（Analysis of composition of microbiomes）
- 是一种基于比值的方法，用于检测不同分组之间的差异物种。它利用W值来衡量物种在不同分组中的相对丰度变化，并使用置换检验来计算显著性水平。
PERMDISP（Permutational multivariate analysis of dispersion）
- 是一种基于方差分析的方法，用于检测不同分组之间的群落离散度是否有显著差异，即组内的距离是否相似。它利用F值或t值来衡量离散度差异，并使用置换检验来计算显著性水平。
ANOSIM（Analysis of similarity）
- 是一种基于秩和检验的方法，用于检测不同分组之间的群落相似度是否有显著差异，即组间的距离是否大于组内的距离。它利用R值来衡量相似度差异，并使用置换检验来计算显著性水平。
PERMANOVA（Permutational multivariate analysis of variance）
- 是一种基于方差分析的方法，用于检测不同分组或环境因子对群落变异的贡献度是否有显著差异，即不同分组能否解释群落变异的程度。它利用F值或R2值来衡量贡献度差异，并使用置换检验来计算显著性水平。
```

```bash
1. PERMANOVA 和 ANOSIM:
原理：

PERMANOVA： 基于样本间的距离矩阵，通过对样本排列进行多重置换来评估组间的整体差异。
ANOSIM： 利用样本间的相似性矩阵，计算R值表示组别间的相异程度。
应用：

用于比较多个组别的整体微生物群落结构是否存在显著差异。
2. CCA 和 RDA:
原理：

CCA 和 RDA： 这两者都是多变量分析方法，通过构建轴来表示解释环境因子对微生物群落结构的影响。CCA强调线性关系，而RDA更加灵活。
应用：

揭示环境因子对微生物群落结构的影响，提供环境因子与微生物的关系图。
3. LEfSe:
原理：

使用线性判别分析（LDA），LEfSe通过考虑微生物特征的丰度和出现频率来找到在不同组别中显著富集或贫乏的特征。
应用：

发现在不同组别中显著富集或贫乏的微生物特征，用于生物学上的解释。
4. ANCOM:
原理：

使用组间的Wilcoxon秩和检验，ANCOM检测微生物群落中相对丰度的差异。
应用：

用于检测微生物群落中的相对丰度的差异，尤其关注于零丰度特征。
5. PERMDISP:
原理：

利用离散度指标，检测每个组别内样本的离散度差异。
应用：

用于评估样本组内的离散度是否存在显著差异，提供关于群落稳定性的信息。
相同之处：
所有这些方法都是用于分析多组样本间的差异，关注于微生物群落结构或功能的差异性。

它们都基于样本的排列或置换进行统计检验，通过构建模型或计算统计量来评估差异的显著性。

差异之处：
应用方向：

PERMANOVA、ANOSIM、CCA、RDA 主要用于整体性结构和环境关系的分析。
LEfSe、ANCOM 主要关注差异特征或相对丰度。
PERMDISP 主要关注样本组内的离散度。
重点内容：

PERMANOVA、ANOSIM、CCA、RDA 重点在于微生物群落整体结构。
LEfSe、ANCOM 重点在于发现差异特征。
PERMDISP 重点在于样本组内的差异性。
数学方法：

PERMANOVA、ANOSIM、CCA、RDA 使用距离矩阵的排列统计方法。
LEfSe 使用线性判别分析（LDA）。
ANCOM 使用组间的Wilcoxon秩和检验。
PERMDISP 使用多变量离散度的排列检验。
通过综合使用这些方法，研究者可以更全面地理解微生物群落的动态变化，从而更深入地解读生态系统的状态和响应。
-------------------
分析顺序：
1. PERMANOVA 或 ANOSIM:
首先使用PERMANOVA或ANOSIM，确定样本组别（条件）之间是否存在整体显著差异。
2. ANCOM:
如果 PERMANOVA 或 ANOSIM 显示存在显著差异，再使用ANCOM检测具体的微生物特征在不同组别中是否有相对丰度的显著差异。
3. PERMDISP:
最后，如果 ANCOM 或 PERMANOVA/ANOSIM 表明存在差异，使用PERMDISP检测样本组内的离散度是否有显著差异，提供有关组内异质性的信息。

这样的分析顺序可以有助于深入理解微生物群落的整体结构差异、特定特征的相对丰度变化以及样本组内的异质性。
```

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 模拟更真实的数据
np.random.seed(42)

# 创建示例数据
samples = ['Sample1', 'Sample2', 'Sample3', 'Sample4']
otu_axis1 = np.random.normal(0, 1, 4)
otu_axis2 = np.random.normal(0, 1, 4)
env_axis1 = np.random.normal(0, 1, 4)
env_axis2 = np.random.normal(0, 1, 4)

data = {'Sample': samples, 'OTU_Axis1': otu_axis1, 'OTU_Axis2': otu_axis2, 'Env_Axis1': env_axis1, 'Env_Axis2': env_axis2}
rda_results = pd.DataFrame(data)

# 绘制OTU和环境因子的RDA分析结果图
plt.figure(figsize=(12, 8))

# 绘制OTU的散点图
sns.scatterplot(x='OTU_Axis1', y='OTU_Axis2', data=rda_results, hue='Sample', palette='viridis', s=100, label='OTU', edgecolor='black', linewidth=1)

# 绘制环境因子的箭头并添加通用注释
arrow_props = dict(facecolor='red', alpha=0.7, width=0.015, head_width=0.05, length_includes_head=True)
for i, row in rda_results.iterrows():
    plt.arrow(0, 0, row['Env_Axis1'], row['Env_Axis2'], **arrow_props)
    
# 添加通用注释
arrow_center = rda_results[['Env_Axis1', 'Env_Axis2']].mean()
plt.annotate('Environment', xy=(arrow_center['Env_Axis1'], arrow_center['Env_Axis2']),
             xytext=(arrow_center['Env_Axis1']+0.1, arrow_center['Env_Axis2']+0.1),
             color='black', arrowprops=dict(facecolor='black', arrowstyle='->', alpha=0.7), fontsize=10, ha='left', va='bottom')

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 添加轴标签和标题
plt.xlabel('OTU Axis 1', fontsize=12)
plt.ylabel('OTU Axis 2', fontsize=12)
plt.title('RDA Analysis of OTU and Environmental Factors', fontsize=14)

# 调整图例位置
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, fontsize=10)

# 显示图形
plt.show()
````

### 3.1.1. <a id='toc3_1_1_'></a>[‰ªéËäÇÁÇπÂíåËæπÂàõÂª∫](#toc0_)

```python
import networkx as nx


# ÂàõÂª∫Á©∫Âõæ
G = nx.Graph()

# Ê∑ªÂä†ËäÇÁÇπ
G.add_node(1)                          # Âçï‰∏™ËäÇÁÇπ
G.add_nodes_from([2, 3])               # Â§ö‰∏™ËäÇÁÇπ
G.add_node("A", role="hub")            # Â∏¶Â±ûÊÄßÁöÑËäÇÁÇπ

# Ê∑ªÂä†Ëæπ
G.add_edge(1, 2)                       # ÂçïÊù°Ëæπ
G.add_edges_from([(1, 3), (2, 3)])     # Â§öÊù°Ëæπ
G.add_edge(1, 2, weight=4.7)           # Â∏¶ÊùÉÈáçÁöÑËæπ

# Âà†Èô§ÂÖÉÁ¥†
G.remove_node(2)                       # Âà†Èô§ËäÇÁÇπÔºàËá™Âä®Âà†Èô§ÂÖ≥ËÅîËæπÔºâ
G.remove_edge(1, 3)                    # Âà†Èô§Ëæπ

print("ËäÇÁÇπÂàóË°®:", G.nodes())          # ËæìÂá∫: [1, 2, 3, 'A']
print("ËæπÂàóË°®:", G.edges())            # ËæìÂá∫: [(1, 2), (1, 3), (2, 3)]
print("ËäÇÁÇπ1ÁöÑÈÇªÂ±Ö:", list(G.neighbors(1)))  # ËæìÂá∫: [2, 3]
print("ÂõæÁöÑÂ∫¶ÂàÜÂ∏É:", nx.degree(G))     # ËæìÂá∫: {1: 2, 2: 2, 3: 2, 'A': 0}

nx.draw(G)
```

```python
def draw_shortest_path(G, pos, shortest_path):
    # nonrelated nodes
    nx.draw_networkx_nodes(G, pos, nodelist=[node for node in G.nodes if node not in shortest_path], node_color='lightblue', node_size=100, edgecolors='black')
    # related nodes
    nx.draw_networkx_nodes(G, pos, nodelist=[node for node in G.nodes if node in shortest_path], node_color='red', node_size=300, edgecolors='black')
    # set node labels
    nx.draw_networkx_labels(G, pos, font_size= 6)
    # all edges
    nx.draw_networkx_edges(G, pos, edge_color= "gray")
    # related edges
    nx.draw_networkx_edges(G, pos, edgelist=[(edge[0], edge[1]) for edge in G.edges if (edge[0] in shortest_path) and (edge[1] in shortest_path)], edge_color='red')
    plt.axis("off")


plt.subplot(1, 2, 1)
# shortest path 
shortest_path = nx.shortest_path(G, source=0, target=33)
print(f"Shortest path from 0 to 33 is : {shortest_path}")
draw_shortest_path(G, pos, shortest_path)

plt.subplot(1, 2, 2)
# shortest path 
shortest_path = nx.shortest_path(G, source=0, target=32)
print(f"Shortest path from 0 to 32 is : {shortest_path}")
draw_shortest_path(G, pos, shortest_path)
```

|Ê°•Ê¢ÅËäÇÁÇπÁ≠õÈÄâ | ‰ªãÊï∞‰∏≠ÂøÉÊÄßÔºàBetweennessÔºâ | ÈÄâtop 5%-10%ÁöÑËäÇÁÇπ | Ê°•Ê¢ÅËßíËâ≤ÔºåÊéßÂà∂‰ø°ÊÅØÊµÅ|
|Ê®°ÂùóÊ†∏ÂøÉÁ≠õÈÄâ | Ê®°ÂùóÂÜÖÂ∫¶ÔºàWithin-Module DegreeÔºâ | È´òÊ®°ÂùóÂÜÖ‰∏≠ÂøÉÊÄßÔºàz-scoreÈ´òÔºâ | Ê®°ÂùóÊ†∏ÂøÉÔºåÁâπÂÆöÂäüËÉΩÊéßÂà∂ËÄÖ|
|Âá∫Â∫¶/ÂÖ•Â∫¶ÊØî‰æãÂàÜÊûêÔºàÊúâÂêëÂõæÔºâ | Âá∫Â∫¶/ÂÖ•Â∫¶ÊØî | Âá∫Â∫¶ ‚â´ ÂÖ•Â∫¶ | ‰∏ªÂØºÂûãËäÇÁÇπÔºàmaster regulatorsÔºâ|
|È´òËÅöÁ±ªËäÇÁÇπÁ≠õÈÄâ | ËÅöÁ±ªÁ≥ªÊï∞ÔºàClusteringÔºâÈ´ò | ËÅöÁ±ªÁ≥ªÊï∞È´ò+Â∫¶ÈÄÇ‰∏≠ | Á®≥ÂÆöÂ≠êÁ≥ªÁªüÁöÑÁª¥Êä§ËÄÖ|

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd


# ÂàõÂª∫‰∏Ä‰∏™Êó†ÂêëÁΩëÁªú
G = nx.Graph()

# Ê∑ªÂä†ËäÇÁÇπÔºàÂÅáËÆæÊòØÂæÆÁîüÁâ©ÂêçÂ≠óÔºâ
microbes = ['Akkermansia', 'Bacteroides', 'Faecalibacterium', 'Clostridium', 'Lactobacillus', 'Escherichia', 'Streptococcus']
G.add_nodes_from(microbes)

# Ê∑ªÂä†ËæπÔºàÂÅáËÆæÂü∫‰∫éÂÖ±Áé∞ÂÖ≥Á≥ªÔºâ
edges = [
    ('Akkermansia', 'Bacteroides'),
    ('Akkermansia', 'Faecalibacterium'),
    ('Bacteroides', 'Clostridium'),
    ('Faecalibacterium', 'Clostridium'),
    ('Faecalibacterium', 'Lactobacillus'),
    ('Lactobacillus', 'Escherichia'),
    ('Escherichia', 'Streptococcus'),
    ('Streptococcus', 'Bacteroides'),
    ('Clostridium', 'Streptococcus')
]
G.add_edges_from(edges)

G = nx.karate_club_graph()

pos = nx.spring_layout(G, seed= 123)
# Áîª‰∏ÄÁîªÁΩëÁªú
nx.draw(G, pos= pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=10)
plt.show()
```

```python
# ËÆ°ÁÆóÊåáÊ†á
degree_dict = dict(G.degree())
betweenness_dict = nx.betweenness_centrality(G)
clustering_dict = nx.clustering(G)

# Êï¥ÁêÜÊàêDataFrame
metrics_df = pd.DataFrame({
    'Degree': degree_dict,
    'Betweenness': betweenness_dict,
    'Clustering': clustering_dict
})

# ÁúãÁúãÁªìÊûú
print(metrics_df)
```

Êàë‰ª¨ËÆæÂÆöÁÆÄÂçïËßÑÂàôÔºö
- DegreeÂâç30%ÁöÑËäÇÁÇπ
- Êàñ BetweennessÂâç30%ÁöÑËäÇÁÇπ

```python
# ËÆ°ÁÆóDegreeÂíåBetweennessÁöÑÈòàÂÄº
degree_threshold = metrics_df['Degree'].quantile(0.7)
betweenness_threshold = metrics_df['Betweenness'].quantile(0.7)

# Á≠õÈÄâÂÖ≥ÈîÆËäÇÁÇπ
key_nodes = metrics_df[
    (metrics_df['Degree'] >= degree_threshold) | 
    (metrics_df['Betweenness'] >= betweenness_threshold)
].index.tolist()

print("Á≠õÈÄâÂá∫ÁöÑÂÖ≥ÈîÆËäÇÁÇπ:", key_nodes)
```

```python
# ÁîªÂõæ
# pos = nx.spring_layout(G, seed= 123)  # Âõ∫ÂÆöÂ∏ÉÂ±Ä
pos = nx.circular_layout(G)

plt.figure(figsize=(8,6))

# ÊôÆÈÄöËäÇÁÇπ
nx.draw_networkx_nodes(G, pos, nodelist=[n for n in G.nodes if n not in key_nodes], node_color='skyblue', node_size=200)

# ÂÖ≥ÈîÆËäÇÁÇπ
nx.draw_networkx_nodes(G, pos, nodelist=key_nodes, node_color='red', node_size=600)

nx.draw_networkx_edges(G, pos)

nx.draw_networkx_labels(G, pos, font_size=10)

plt.title("Microbial Network with Key Nodes Highlighted")
plt.axis("off")
```

## 4.3. <a id='toc4_3_'></a>[ML to predict the keystone special](#toc0_)

### 4.3.1. <a id='toc4_3_1_'></a>[ÁâπÂæÅÊûÑÂª∫](#toc0_)

```python
import networkx as nx
import pandas as pd


G = nx.karate_club_graph()  # Á§∫‰æãÂ∞èÁΩëÁªúÔºåÂêéÈù¢ÂèØ‰ª•Êç¢ÂæÆÁîüÁâ©ÁΩëÁªú

# ÁâπÂæÅÊèêÂèñ
features = pd.DataFrame({
    'Degree': dict(G.degree()),
    'Betweenness': nx.betweenness_centrality(G),
    'Closeness': nx.closeness_centrality(G),
    'Clustering': nx.clustering(G),
    'Average_Neighbor_Degree': nx.average_neighbor_degree(G)
})

features.head()
```

### 4.3.2. <a id='toc4_3_2_'></a>[Êó†ÁõëÁù£ÂºÇÂ∏∏Ê£ÄÊµãÔºàÊØîÂ¶ÇIsolation ForestÊâæÊúÄÂºÇÂ∏∏ËäÇÁÇπÔºâ](#toc0_)

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.1, random_state=42)
features['Anomaly_Score'] = iso_forest.fit_predict(features)

# Âè™‰øùÁïôÂºÇÂ∏∏ËäÇÁÇπÔºàÈ¢ÑÊµãÂÄº‰∏∫-1Ôºâ
anomalous_nodes = features[features['Anomaly_Score'] == -1]
print("ÂºÇÂ∏∏ÂÖ≥ÈîÆËäÇÁÇπÔºö", anomalous_nodes.index.tolist())
```

### 4.3.3. <a id='toc4_3_3_'></a>[ÁâπÂæÅÈáçË¶ÅÊÄßÊéíÂ∫èÔºàÊØîÂ¶ÇÁî®XGBoostÔºâ](#toc0_)

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

# ‰∫∫‰∏∫ÈÄ†‰∏Ä‰∏™Ê†áÁ≠æÔºåÂÅáËÆæÂâçÂá†‰∏™ËäÇÁÇπÊòØÈáçË¶ÅÔºàÁúüÂÆûÊï∞ÊçÆ‰Ω†ÂèØ‰ª•Áî®ÂäüËÉΩÊ≥®ÈáäÊàñËÄÖÁîüÁâ©Áü•ËØÜÔºâ
features['Label'] = 0
features.loc[anomalous_nodes.index, 'Label'] = 1

X = features.drop(columns=['Anomaly_Score', 'Label'])
y = features['Label']

# Âª∫Ê®°
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X, y)

# ËæìÂá∫ÁâπÂæÅÈáçË¶ÅÊÄß
importances = model.feature_importances_
importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances}).sort_values('Importance', ascending=False)
print(importance_df)
```

### 4.4.1. <a id='toc4_4_1_'></a>[Ê†∏ÂøÉÊÄùË∑Ø](#toc0_)

ÂéüÊù•Êàë‰ª¨ÂàÜÊûêÁöÑÊòØ‰∏ÄÂº†ÈùôÊÄÅÂõæÔºåÁé∞Âú®ÊòØÔºö

- ‰∏çÂêåÊó∂Èó¥ÁÇπÔºåÊØè‰∏™Êó∂Èó¥ÁÇπ‰∏ÄÂº†ÁΩëÁªúÂõæÔºåËäÇÁÇπÂíåËæπÈÉΩÂèØËÉΩÂèòÂåñÔºÅ

‰∫éÊòØÊàë‰ª¨ÂèØ‰ª•Ôºö

- ÊØîËæÉËäÇÁÇπÂú®‰∏çÂêåÊó∂Èó¥ÁöÑÂõæÊåáÊ†áÂèòÂåñ

- ËØÜÂà´Âá∫ÊåáÊ†áÂâßÁÉàÂèòÂåñÁöÑËäÇÁÇπÔºàÂä®ÊÄÅÈáçË¶ÅËäÇÁÇπÔºâ

- ËøòÂèØ‰ª•Âª∫**Âä®ÊÄÅÂõæÁ•ûÁªèÁΩëÁªúÔºàDynamic GNNÔºâ**Êù•Áõ¥Êé•Âª∫Ê®°ÊºîÂèòËøáÁ®ãÔºÅÔºàÈ´òÁ∫ßÁâàÔºâ


„Äê‰∏çÂêåÊó∂Èó¥ÈááÊ†∑Êï∞ÊçÆ„Äë‚Üí„ÄêÊûÑÂª∫ÊØè‰∏™Êó∂Èó¥ÁÇπÁöÑÁΩëÁªú„Äë‚Üí„ÄêËÆ°ÁÆóÊØè‰∏™ËäÇÁÇπÂêÑÊó∂Èó¥ÁÇπÁöÑÂõæÊåáÊ†á„Äë‚Üí„ÄêÂàÜÊûêÊåáÊ†áÂèòÂåñË∂ãÂäø„Äë‚Üí„ÄêÁ≠õÈÄâÊåáÊ†áÂèòÂåñÊúÄÂ§ßÁöÑËäÇÁÇπ„Äë


```python
import networkx as nx


# Êó∂Èó¥ÁÇπ0
G0 = nx.Graph()
G0.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D')])

# Êó∂Èó¥ÁÇπ1
G1 = nx.Graph()
G1.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E')])

# Êó∂Èó¥ÁÇπ2
G2 = nx.Graph()
G2.add_edges_from([('C', 'D'), ('D', 'E'), ('E', 'F'), ('F', 'G')])

# ÊääÊó∂Èó¥Â∫èÂàóÁΩëÁªúÊîæÂú®ÂàóË°®Èáå
networks = [G0, G1, G2]
```

```python
plt.figure()
for i, G in enumerate(networks):
    plt.subplot(1, 3, i + 1)
    pos = nx.spring_layout(G, seed=123)
    nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', edgecolors='black')
    plt.title(f"Time Point {i}")
plt.tight_layout()
```

ÊèêÂèñÊØè‰∏™ËäÇÁÇπÊØè‰∏™Êó∂Èó¥ÁÇπÁöÑÊåáÊ†á

```python
import pandas as pd

# Â≠òÂÇ®ÊåáÊ†á
metrics_over_time = {}

for t, G in enumerate(networks):
    degree = dict(G.degree())
    betweenness = nx.betweenness_centrality(G)
    
    for node in G.nodes:
        if node not in metrics_over_time:
            metrics_over_time[node] = []
        metrics_over_time[node].append({
            'time': t,
            'degree': degree.get(node, 0),
            'betweenness': betweenness.get(node, 0)
        })

# Êï¥ÁêÜÊàêDataFrame
records = []
for node, metrics in metrics_over_time.items():
    for m in metrics:
        records.append({'Node': node, 'Time': m['time'], 'Degree': m['degree'], 'Betweenness': m['betweenness']})

df = pd.DataFrame(records)
print(df)
```

ËÆ°ÁÆóÂèòÂåñÂπÖÂ∫¶
- ÊØîÂ¶ÇÔºåÁÆóÂá∫ÊØè‰∏™ËäÇÁÇπDegreeÈöèÊó∂Èó¥ÂèòÂåñÁöÑ‚ÄúÊÄªÂèòÂåñÂπÖÂ∫¶‚ÄùÔºö

```python
# ËÆ°ÁÆóDegreeÂèòÂåñ
pivot_degree = df.pivot(index='Node', columns='Time', values='Degree').fillna(0)
pivot_betweenness = df.pivot(index='Node', columns='Time', values='Betweenness').fillna(0)

# ÂèòÂåñÈáèÔºàÂèñÂ∑ÆÁöÑÁªùÂØπÂÄºÔºâ
degree_change = pivot_degree.diff(axis=1).abs().sum(axis=1)
betweenness_change = pivot_betweenness.diff(axis=1).abs().sum(axis=1)

# ÂêàÂπ∂Êàê‰∏Ä‰∏™Ë°®
change_summary = pd.DataFrame({
    'Degree_Total_Change': degree_change,
    'Betweenness_Total_Change': betweenness_change
}).sort_values('Degree_Total_Change', ascending=False)

print(change_summary)
```

```python
# ÂèñDegreeÂèòÂåñÊúÄÂ§ßÁöÑËäÇÁÇπ
key_nodes_degree = change_summary.sort_values('Degree_Total_Change', ascending=False).head(3).index.tolist()
# ÂèñBetweennessÂèòÂåñÊúÄÂ§ßÁöÑËäÇÁÇπ
key_nodes_betweenness = change_summary.sort_values('Betweenness_Total_Change', ascending=False).head(3).index.tolist()

print("DegreeÂä®ÊÄÅÂÖ≥ÈîÆËäÇÁÇπ:", key_nodes_degree)
print("BetweennessÂä®ÊÄÅÂÖ≥ÈîÆËäÇÁÇπ:", key_nodes_betweenness)
```

```python
def set_edge_color(
        G,  
        vertexes_list:list, 
        default_vertex:dict= {"size":12, "color":"lightblue"}, 
        highlight_vertex:int= {"size":15, "color":"red"},
) -> tuple:
    vertex_size_list = []
    vertex_color_list = []
    edge_color_list = []
    # For vertex
    for vertex in G.vs:
        if vertex['name'] in vertexes_list:
            vertex_size_list.append(highlight_vertex['size'])
            vertex_color_list.append(highlight_vertex['color'])
        else:
            vertex_size_list.append(default_vertex['size'])
            vertex_color_list.append(default_vertex['color'])
    # For edge
    for edge in G.es:
        if (edge.source in vertexes_list) and (edge.target in vertexes_list): # choosed
            edge_color_list.append("red")
        else:
            edge_color_list.append("gray")
            
    return (vertex_size_list, vertex_color_list, edge_color_list)


# Highlight some vertexes and edges
vertex_size_list, vertex_color_list, edge_color_list = set_edge_color(G= G, vertexes_list= [0, 8, 32])



# draw the picture
fig, ax = plt.subplots(figsize= (4, 4))
ig.plot(
    G,
    layout= G.layout("circle"),
    bbox= (300, 300),
    target= ax,

    # vertex
    vertex_size= vertex_size_list,
    vertex_color= vertex_color_list,

    # vertex label
    # vertex_label= ["demo"] *34,
    vertex_label_size= 5,
    vertex_label_color= 'black',

    # edge
    edge_color= edge_color_list,

    # dege label
)
plt.title("karate club", fontsize= 12)
plt.tight_layout()
```

#### 5.2.4.1. <a id='toc5_2_4_1_'></a>[ËäÇÁÇπÂ±ÇÈù¢ÊåáÊ†áÔºàNode-level MetricsÔºâ](#toc0_)

```python
# ËäÇÁÇπÂ∫¶ Degree
# ÂÆö‰πâÔºöÊüêËäÇÁÇπËøûÊé•ÁöÑËæπÁöÑÊï∞ÈáèÔºådeg(v) = #edges connected to v
# ÊÑè‰πâÔºöË°°ÈáèËäÇÁÇπÁöÑÁõ¥Êé•ËøûÊé•ËÉΩÂäõÊàñÊú¨Âú∞ÂΩ±ÂìçÂäõ
degree = g.degree()

# ‰ªãÊï∞‰∏≠ÂøÉÊÄß Betweenness Centrality
# ÂÆö‰πâÔºöC_B(v) = Œ£_{s‚â†v‚â†t} (œÉ_st(v) / œÉ_st)
# ÊÑè‰πâÔºöËäÇÁÇπ‰Ωú‰∏∫‰ø°ÊÅØÊµÅÁöÑ‚Äú‰∏≠‰ªã‚ÄùÁ®ãÂ∫¶
betweenness = g.betweenness()

# Êé•Ëøë‰∏≠ÂøÉÊÄß Closeness Centrality
# ÂÆö‰πâÔºöC_C(v) = 1 / Œ£ d(v, t)
# ÊÑè‰πâÔºöË°®Á§∫‰∏Ä‰∏™ËäÇÁÇπÂú®ÁΩëÁªú‰∏≠‰∏éÂÖ∂‰ªñËäÇÁÇπÁöÑÂπ≥ÂùáË∑ùÁ¶ªÁöÑÂÄíÊï∞Ôºå‰∏≠ÂøÉÁ®ãÂ∫¶Ë∂äÈ´òÔºåË∑ùÁ¶ªÂÖ∂‰ªñËäÇÁÇπË∂äËøë
closeness = g.closeness()

# ÁâπÂæÅÂêëÈáè‰∏≠ÂøÉÊÄß Eigenvector Centrality
# ÂÆö‰πâÔºöC_E(v) ‚àù Œ£_{u ‚àà N(v)} C_E(u)
# ÊÑè‰πâÔºö‰∏ç‰ªÖËÄÉËôëËøûÊé•Êï∞ÔºåËøòËÄÉËôëËøûÊé•ÂØπË±°ÁöÑ‚ÄúÈáçË¶ÅÊÄß‚ÄùÔºåÈÄÇÂêàË°°Èáè‚ÄúÂ£∞Êúõ‚ÄùÊàñ‚Äú‰º†Êí≠Âäõ‚Äù
eigenvector = g.eigenvector_centrality()

# PageRank
# ÂÆö‰πâÔºöPageRank(v) = (1 - d)/N + d * Œ£ PageRank(u)/deg(u)  over neighbors u
# ÊÑè‰πâÔºöÂü∫‰∫éÈöèÊú∫Ê∏∏Ëµ∞ÁöÑ‰∏≠ÂøÉÊÄßÊåáÊ†áÔºåGoogleÊúÄÊó©Áî®‰∫éÁΩëÈ°µÊéíÂêç
pagerank = g.pagerank()

# Â±ÄÈÉ®ËÅöÁ±ªÁ≥ªÊï∞ Clustering Coefficient
# ÂÆö‰πâÔºöC(v) = 2 * (# of links between neighbors of v) / (deg(v) * (deg(v)-1))
# ÊÑè‰πâÔºöË°°ÈáèÈÇªÂ±Ö‰πãÈó¥ÁöÑËøûÊé•Á®ãÂ∫¶ÔºàÂ±ÄÈÉ®Âõ¢ÁªìÊûÑÔºâ
clustering = g.transitivity_local_undirected(mode="zero")

# ÊâìÂåÖ‰∏∫ DataFrame
df_nodes = pd.DataFrame({
    "Node": range(g.vcount()),
    "Degree": degree,
    "Betweenness": betweenness,
    "Closeness": closeness,
    "Eigenvector": eigenvector,
    "PageRank": pagerank,
    "ClusteringCoeff": clustering
})

df_nodes
```

#### 5.2.4.2. <a id='toc5_2_4_2_'></a>[ËæπÂ±ÇÈù¢ÊåáÊ†áÔºàEdge-level MetricsÔºâ](#toc0_)

```python
# Ëæπ‰ªãÊï∞‰∏≠ÂøÉÊÄß Edge Betweenness
# ÂÆö‰πâÔºöC_B(e) = Œ£_{s‚â†t} (œÉ_st(e) / œÉ_st)
# ÊÑè‰πâÔºöË°°ÈáèÊüê‰∏ÄËæπÂú®ËøûÊé•‰∏çÂêåÁΩëÁªúÂå∫ÂüüÊó∂ÁöÑ‚ÄúÊ°•Ê¢Å‰ΩúÁî®‚Äù
edge_betweenness = g.edge_betweenness()
g.es["betweenness"] = edge_betweenness

# ËæìÂá∫Ââç5Êù°ËæπÁöÑÊåáÊ†áÂÄº
for i, e in enumerate(g.es[:5]):
    print(f"Edge {e.tuple}: Betweenness = {e['betweenness']:.2f}")
```

## 5.3. <a id='toc5_3_'></a>[Screening the keystone taxa](#toc0_)

ÈÄöÂ∏∏‰Ω†Â∑≤ÁªèÊúâ‰∏Ä‰∏™Áõ∏ÂÖ≥ÊÄßÁü©ÈòµÔºàÂ¶ÇSparCC, Spearman, PearsonÔºâÔºåËøôÈáåÊàë‰∏æÁÆÄÂçï‰æãÂ≠êÔºö

```python
import igraph as ig
import numpy as np


# Á§∫‰æãÔºö10‰∏™ÂæÆÁîüÁâ©
adj_matrix = np.random.rand(10, 10)
adj_matrix = (adj_matrix + adj_matrix.T) / 2  # ÂØπÁß∞Âåñ
adj_matrix[adj_matrix < 0.7] = 0              # Âè™‰øùÁïôÂº∫Áõ∏ÂÖ≥

g = ig.Graph.Weighted_Adjacency(adj_matrix.tolist(), mode="undirected", attr="weight", loops=False)
```

ËÆ°ÁÆóÊ†∏ÂøÉÊåáÊ†á:

```python
# ËÆ°ÁÆóÂêÑËäÇÁÇπÊåáÊ†á
degree = g.degree()
betweenness = g.betweenness()
closeness = g.closeness()
eigenvector = g.eigenvector_centrality()
```

ËµãÂÄºÂà∞ËäÇÁÇπÂ±ûÊÄßÔºå‰æø‰∫éÂêéÈù¢Á≠õÈÄâÔºö

```python
g.vs["degree"] = degree
g.vs["betweenness"] = betweenness
g.vs["closeness"] = closeness
g.vs["eigenvector"] = eigenvector
```

ÈÄöÂ∏∏ÁöÑÊ†áÂáÜÂèØ‰ª•ÊòØÔºö
- Â∫¶Êï∞ > Âπ≥ÂùáÂ∫¶
- ‰ªãÊï∞‰∏≠ÂøÉÊÄßÂâç 10%
- ÁâπÂæÅÂêëÈáè‰∏≠ÂøÉÊÄßÈ´ò
- ËøûÂà∞Â§ö‰∏™Ê®°Âùó

core_microbes ÊòØÁ≠õÈÄâÂêéÁöÑËäÇÁÇπÂ≠êÈõÜ„ÄÇ

```python
import numpy as np


threshold = np.mean(degree)

core_microbes = g.vs.select(degree_gt=threshold)
```

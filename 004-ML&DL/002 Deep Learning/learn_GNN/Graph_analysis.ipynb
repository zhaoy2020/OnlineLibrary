## 1.1. <a id='toc1_1_'></a>[什么是图？](#toc0_)

- 图的关键要素包括：顶点（vertex）和边（edge）。

- 数学上一般把图记为，G=(V,E)其中 𝑉 是节点集合，𝐸是边集合。

- 图可以分成很多种类型：

  - 有向图（Directed Graph）/无向图（Undirected Graph）

  - 无权/加权图（Weighted Graph）：边有权重，比如A对B影响的强度是0.8。

  - 稀疏图/稠密图（Sparse/Dense Graph）：边比较少还是比较多。

# 3. <a id='toc3_'></a>[Networkx](#toc0_)

- 图的基本组成
  - 节点（Node）：可以是任意哈希对象（数字、字符串、自定义对象）
  - 边（Edge）：连接两个节点的关系，可携带属性（如权重、类型）

- 图（Graph）类型
    |类型	|创建方式	|特点|
    |-|-|-|
    |<span style="color:red">无向图</span>	|G = nx.Graph()	|边无方向，适用于社交网络、交通网|
    |有向图	|G = nx.DiGraph()	|边有方向，适用于网页链接、工作流|
    |多重图	|G = nx.MultiGraph()	|允许节点间存在多条边|
    |加权图	|添加weight属性	|边带权重，用于表示距离、成本等|




## 3.1. <a id='toc3_1_'></a>[创建与修改图](#toc0_)

### 3.1.1. <a id='toc3_1_1_'></a>[从节点和边创建](#toc0_)

```python
import networkx as nx


# 创建空图
G = nx.Graph()

# 添加节点
G.add_node(1)                          # 单个节点
G.add_nodes_from([2, 3])               # 多个节点
G.add_node("A", role="hub")            # 带属性的节点

# 添加边
G.add_edge(1, 2)                       # 单条边
G.add_edges_from([(1, 3), (2, 3)])     # 多条边
G.add_edge(1, 2, weight=4.7)           # 带权重的边

# 删除元素
G.remove_node(2)                       # 删除节点（自动删除关联边）
G.remove_edge(1, 3)                    # 删除边

print("节点列表:", G.nodes())          # 输出: [1, 2, 3, 'A']
print("边列表:", G.edges())            # 输出: [(1, 2), (1, 3), (2, 3)]
print("节点1的邻居:", list(G.neighbors(1)))  # 输出: [2, 3]
print("图的度分布:", nx.degree(G))     # 输出: {1: 2, 2: 2, 3: 2, 'A': 0}

nx.draw(G)
```

### 3.5.3. <a id='toc3_5_3_'></a>[度](#toc0_)

- 高入度节点：被很多基因或微生物调控，通常是受控目标（如核心代谢基因）。

## 3.6. <a id='toc3_6_'></a>[Path](#toc0_)

nx.shortest_path(G, source=0, target=33)

```python
def draw_shortest_path(G, pos, shortest_path):
    # nonrelated nodes
    nx.draw_networkx_nodes(G, pos, nodelist=[node for node in G.nodes if node not in shortest_path], node_color='lightblue', node_size=100, edgecolors='black')
    # related nodes
    nx.draw_networkx_nodes(G, pos, nodelist=[node for node in G.nodes if node in shortest_path], node_color='red', node_size=300, edgecolors='black')
    # set node labels
    nx.draw_networkx_labels(G, pos, font_size= 6)
    # all edges
    nx.draw_networkx_edges(G, pos, edge_color= "gray")
    # related edges
    nx.draw_networkx_edges(G, pos, edgelist=[(edge[0], edge[1]) for edge in G.edges if (edge[0] in shortest_path) and (edge[1] in shortest_path)], edge_color='red')
    plt.axis("off")


plt.subplot(1, 2, 1)
# shortest path 
shortest_path = nx.shortest_path(G, source=0, target=33)
print(f"Shortest path from 0 to 33 is : {shortest_path}")
draw_shortest_path(G, pos, shortest_path)

plt.subplot(1, 2, 2)
# shortest path 
shortest_path = nx.shortest_path(G, source=0, target=32)
print(f"Shortest path from 0 to 32 is : {shortest_path}")
draw_shortest_path(G, pos, shortest_path)
```

### 3.6.1. <a id='toc3_6_1_'></a>[路径长度（Path Length）和平均路径长度（APL）](#toc0_)

数学原理
- 两个节点间的最短路径上的边数。
- 全网的平均路径长度：

生物学解释
- 短路径：系统响应快（如环境变化，基因快速调整表达）。
- 长路径：调控级联较多，可能有复杂中间层（如代谢多级反应链）。

例子
- 在抗生素应激下，E. coli可能通过一两步快速调控耐药相关基因，体现低APL特征。

```python
# 单对节点的最短路径长度
length = nx.shortest_path_length(G, source=1, target=5)
print("节点1到节点5的路径长度:", length)
```

```python
# 整个图的平均路径长度
apl = nx.average_shortest_path_length(G)
print("图的平均路径长度（APL）:", apl)
```

### 3.6.2. <a id='toc3_6_2_'></a>[网络直径（Diameter）](#toc0_)

数学原理
- 网络中<span style="color:red">最远两个节点</span>之间的<span style="color:red">最短</span>路径长度。

生物学解释
- 描述系统最极端情况下的最大调控延迟。
- 微生物群体中，直径较小意味着生态联系紧密，信息传播快。

图必须是连通的（即任何两个节点之间都能到达）

```python
# 计算直径
diameter = nx.diameter(G)
print("图的直径是:", diameter)
```

如果图是不连通的，需要提取最大连通子图再算：

```python
# 提取最大连通子图
largest_cc = max(nx.connected_components(G), key=len)
G_lcc = G.subgraph(largest_cc)

# 计算直径
diameter = nx.diameter(G_lcc)
print("最大连通子图的直径是:", diameter)
```

## 3.7. <a id='toc3_7_'></a>[中心指标](#toc0_)

评估节点在网络中的重要性，间接反映网络复杂性：

### 3.7.1. <a id='toc3_7_1_'></a>[介数中心性（Betweenness Centrality）](#toc0_)

数学原理
- 衡量`一个节点`在网络中作为其他节点之间桥梁的重要性。

物学解释
- 高介数节点：关键中枢控制点，控制信息/物质流动。
- 微生物生态中，这类微生物一旦消失，生态系统极易崩溃（称为Keystone Species关键物种）。

例子
- 在肠道菌群中，部分低丰度微生物（如Faecalibacterium prausnitzii）拥有超高介数中心性，对肠道健康极重要。

```python
# 介数中心性（节点版）
centrality = nx.betweenness_centrality(G)

# 介数中心性（边版）,（是的，边也有介数中心性！）
centraliby_edge = nx.edge_betweenness_centrality(G)

# 打印结果
for node, value in {t[0]:t[1] for t in sorted(centrality.items(), key=lambda d:d[1], reverse=True)}.items():
    print(f"节点 {node} 的介数中心性: {value:.4f}")
```

### 3.7.2. <a id='toc3_7_2_'></a>[接近中心性（Closeness Centrality）](#toc0_)

Graphcal metric
- 一个节点到所有其他节点的距离有多短。
- 节点越容易（通过短路径）接触到网络中其他节点，它的接近中心性就越高。

简单理解：
- "一个节点离大家都比较近吗？近的话，它的接近中心性就高。"

```python
closeness = nx.closeness_centrality(G)

# 打印结果
for node, value in {t[0]:t[1] for t in sorted(closeness.items(), key=lambda d:d[1], reverse=True)}.items():
    print(f"节点 {node} 的接近中心性: {value:.4f}")
```

### 3.7.3. <a id='toc3_7_3_'></a>[度中心性（Degree Centrality）](#toc0_)

Graphcal metric
- 度（Degree）：节点直接连接的邻居节点数量。
- 度中心性（Degree Centrality）：一个节点的度数占网络最大可能度数的比例。

简单理解："一个节点连得越多，度中心性越高，它在网络中越重要。"

```python
# 计算度中心性
deg_centrality = nx.degree_centrality(G)

# 打印每个节点的度中心性
for node, value in {t[0]:t[1] for t in sorted(deg_centrality.items(), key=lambda d:d[1], reverse=True)}.items():
    print(f"节点 {node} 的度中心性: {value:.4f}")
```

## 3.8. <a id='toc3_8_'></a>[模块化与聚类特征](#toc0_)

### 3.8.1. <a id='toc3_8_1_'></a>[模块性（Modularity）](#toc0_)

- Graphical metric:
  - 衡量网络分解为模块（子群落）的程度，值越高说明模块结构越明显（通常0.3-0.7为合理范围）。
  - 衡量网络能否自然分成多个子群体（模块），且模块内连接密集，模块间连接稀疏。such as 公式（Louvain modularity简化版）
- Biologcial annotation:
  - 高模块性：存在功能子系统，例如不同代谢功能的基因模块。
- Example:
  - 微生物网络往往展现出模块化，比如产氢菌、产酸菌分别构成功能模块。

如何获得社区？
- 常见的自动社区划分算法有：
  - greedy_modularity_communities(G)：贪心法（适合中小图）
  - label_propagation_communities(G)：标签传播（适合大图）
  - asyn_lpa_communities(G)：异步标签传播
  - louvain（需要安装额外包，比如 python-louvain）

- 手动分社区也可以算模块性：只要你给出分组就行！

```python
from networkx.algorithms.community import greedy_modularity_communities, modularity


# 1. 自动划分社区（贪心算法）
communities = list(greedy_modularity_communities(G))

# 2. 计算模块性
modularity_value = modularity(G, communities)

print("划分出的社区数量:", len(communities))
print("模块性Q值:", modularity_value)
```

## 4.2. <a id='toc4_2_'></a>[基于图指标筛选关键基因/微生物的方法](#toc0_)

核心思路
- 高中心性节点 → 关键调控者 / 关键物种

- 高连通度节点 → 核心基因 / 核心成员

- 桥梁节点（高介数中心性） → 系统稳定性的守护者

- 功能模块中心节点 → 特定生物过程的枢纽

简单来说，就是挑指标高的、位置特殊的节点，因为它们对整个网络或者特定功能最重要！

常见的筛选标准（实际应用中常用）

|策略 | 指标 | 筛选规则 | 结果含义|
|-|-|-|-|
|连通度筛选 | 度（Degree） | 排名前10%节点 | 超级连接者（hubs），控制或被控制|
|桥梁节点筛选 | 介数中心性（Betweenness） | 选top 5%-10%的节点 | 桥梁角色，控制信息流|
|模块核心筛选 | 模块内度（Within-Module Degree） | 高模块内中心性（z-score高） | 模块核心，特定功能控制者|
|出度/入度比例分析（有向图） | 出度/入度比 | 出度 ≫ 入度 | 主导型节点（master regulators）|
|高聚类节点筛选 | 聚类系数（Clustering）高 | 聚类系数高+度适中 | 稳定子系统的维护者|

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd


# 创建一个无向网络
G = nx.Graph()

# 添加节点（假设是微生物名字）
microbes = ['Akkermansia', 'Bacteroides', 'Faecalibacterium', 'Clostridium', 'Lactobacillus', 'Escherichia', 'Streptococcus']
G.add_nodes_from(microbes)

# 添加边（假设基于共现关系）
edges = [
    ('Akkermansia', 'Bacteroides'),
    ('Akkermansia', 'Faecalibacterium'),
    ('Bacteroides', 'Clostridium'),
    ('Faecalibacterium', 'Clostridium'),
    ('Faecalibacterium', 'Lactobacillus'),
    ('Lactobacillus', 'Escherichia'),
    ('Escherichia', 'Streptococcus'),
    ('Streptococcus', 'Bacteroides'),
    ('Clostridium', 'Streptococcus')
]
G.add_edges_from(edges)

G = nx.karate_club_graph()

pos = nx.spring_layout(G, seed= 123)
# 画一画网络
nx.draw(G, pos= pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=10)
plt.show()
```

```python
# 计算指标
degree_dict = dict(G.degree())
betweenness_dict = nx.betweenness_centrality(G)
clustering_dict = nx.clustering(G)

# 整理成DataFrame
metrics_df = pd.DataFrame({
    'Degree': degree_dict,
    'Betweenness': betweenness_dict,
    'Clustering': clustering_dict
})

# 看看结果
print(metrics_df)
```

我们设定简单规则：
- Degree前30%的节点
- 或 Betweenness前30%的节点

```python
# 计算Degree和Betweenness的阈值
degree_threshold = metrics_df['Degree'].quantile(0.7)
betweenness_threshold = metrics_df['Betweenness'].quantile(0.7)

# 筛选关键节点
key_nodes = metrics_df[
    (metrics_df['Degree'] >= degree_threshold) | 
    (metrics_df['Betweenness'] >= betweenness_threshold)
].index.tolist()

print("筛选出的关键节点:", key_nodes)
```

```python
# 画图
# pos = nx.spring_layout(G, seed= 123)  # 固定布局
pos = nx.circular_layout(G)

plt.figure(figsize=(8,6))

# 普通节点
nx.draw_networkx_nodes(G, pos, nodelist=[n for n in G.nodes if n not in key_nodes], node_color='skyblue', node_size=200)

# 关键节点
nx.draw_networkx_nodes(G, pos, nodelist=key_nodes, node_color='red', node_size=600)

nx.draw_networkx_edges(G, pos)

nx.draw_networkx_labels(G, pos, font_size=10)

plt.title("Microbial Network with Key Nodes Highlighted")
plt.axis("off")
```

## 4.4. <a id='toc4_4_'></a>[结合时间序列，动态筛选](#toc0_)

### 4.4.1. <a id='toc4_4_1_'></a>[核心思路](#toc0_)

原来我们分析的是一张静态图，现在是：

- 不同时间点，每个时间点一张网络图，节点和边都可能变化！

于是我们可以：

- 比较节点在不同时间的图指标变化

- 识别出指标剧烈变化的节点（动态重要节点）

- 还可以建**动态图神经网络（Dynamic GNN）**来直接建模演变过程！（高级版）


【不同时间采样数据】→【构建每个时间点的网络】→【计算每个节点各时间点的图指标】→【分析指标变化趋势】→【筛选指标变化最大的节点】


```python
import networkx as nx


# 时间点0
G0 = nx.Graph()
G0.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D')])

# 时间点1
G1 = nx.Graph()
G1.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E')])

# 时间点2
G2 = nx.Graph()
G2.add_edges_from([('C', 'D'), ('D', 'E'), ('E', 'F'), ('F', 'G')])

# 把时间序列网络放在列表里
networks = [G0, G1, G2]
```

```python
plt.figure()
for i, G in enumerate(networks):
    plt.subplot(1, 3, i + 1)
    pos = nx.spring_layout(G, seed=123)
    nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', edgecolors='black')
    plt.title(f"Time Point {i}")
plt.tight_layout()
```

提取每个节点每个时间点的指标

```python
import pandas as pd

# 存储指标
metrics_over_time = {}

for t, G in enumerate(networks):
    degree = dict(G.degree())
    betweenness = nx.betweenness_centrality(G)
    
    for node in G.nodes:
        if node not in metrics_over_time:
            metrics_over_time[node] = []
        metrics_over_time[node].append({
            'time': t,
            'degree': degree.get(node, 0),
            'betweenness': betweenness.get(node, 0)
        })

# 整理成DataFrame
records = []
for node, metrics in metrics_over_time.items():
    for m in metrics:
        records.append({'Node': node, 'Time': m['time'], 'Degree': m['degree'], 'Betweenness': m['betweenness']})

df = pd.DataFrame(records)
print(df)
```

计算变化幅度
- 比如，算出每个节点Degree随时间变化的“总变化幅度”：

```python
# 计算Degree变化
pivot_degree = df.pivot(index='Node', columns='Time', values='Degree').fillna(0)
pivot_betweenness = df.pivot(index='Node', columns='Time', values='Betweenness').fillna(0)

# 变化量（取差的绝对值）
degree_change = pivot_degree.diff(axis=1).abs().sum(axis=1)
betweenness_change = pivot_betweenness.diff(axis=1).abs().sum(axis=1)

# 合并成一个表
change_summary = pd.DataFrame({
    'Degree_Total_Change': degree_change,
    'Betweenness_Total_Change': betweenness_change
}).sort_values('Degree_Total_Change', ascending=False)

print(change_summary)
```

筛选动态关键节点
- 比如，我们取Degree变化最大、或Betweenness变化最大的Top节点：

```python
# 取Degree变化最大的节点
key_nodes_degree = change_summary.sort_values('Degree_Total_Change', ascending=False).head(3).index.tolist()
# 取Betweenness变化最大的节点
key_nodes_betweenness = change_summary.sort_values('Betweenness_Total_Change', ascending=False).head(3).index.tolist()

print("Degree动态关键节点:", key_nodes_degree)
print("Betweenness动态关键节点:", key_nodes_betweenness)
```

画动态变化曲线
- 画一下节点的Degree随时间变化趋势（很直观！）

```python
import matplotlib.pyplot as plt

# 画Degree变化
plt.figure(figsize=(10,6))

for node in key_nodes_degree:
    plt.plot(pivot_degree.columns, pivot_degree.loc[node], label=f'Node {node}')

plt.xlabel('Time')
plt.ylabel('Degree')
plt.title('Degree Changes Over Time')
plt.legend()
```

```python
def set_edge_color(
        G,  
        vertexes_list:list, 
        default_vertex:dict= {"size":12, "color":"lightblue"}, 
        highlight_vertex:int= {"size":15, "color":"red"},
) -> tuple:
    vertex_size_list = []
    vertex_color_list = []
    edge_color_list = []
    # For vertex
    for vertex in G.vs:
        if vertex['name'] in vertexes_list:
            vertex_size_list.append(highlight_vertex['size'])
            vertex_color_list.append(highlight_vertex['color'])
        else:
            vertex_size_list.append(default_vertex['size'])
            vertex_color_list.append(default_vertex['color'])
    # For edge
    for edge in G.es:
        if (edge.source in vertexes_list) and (edge.target in vertexes_list): # choosed
            edge_color_list.append("red")
        else:
            edge_color_list.append("gray")
            
    return (vertex_size_list, vertex_color_list, edge_color_list)


# Highlight some vertexes and edges
vertex_size_list, vertex_color_list, edge_color_list = set_edge_color(G= G, vertexes_list= [0, 8, 32])



# draw the picture
fig, ax = plt.subplots(figsize= (4, 4))
ig.plot(
    G,
    layout= G.layout("circle"),
    bbox= (300, 300),
    target= ax,

    # vertex
    vertex_size= vertex_size_list,
    vertex_color= vertex_color_list,

    # vertex label
    # vertex_label= ["demo"] *34,
    vertex_label_size= 5,
    vertex_label_color= 'black',

    # edge
    edge_color= edge_color_list,

    # dege label
)
plt.title("karate club", fontsize= 12)
plt.tight_layout()
```

| 层级   | 指标                     | 数学符号/定义                          | 意义               |

```python
import igraph as ig
import matplotlib.pyplot as plt
import pandas as pd


g = ig.Graph.Famous("Zachary")

layout = g.layout("fr")  # Fruchterman-Reingold 布局

fig, ax = plt.subplots()
ig.plot(
    g,
    layout=layout,
    vertex_label=g.vs.indices,
    vertex_color="lightblue",
    vertex_size=20,
    edge_arrow_size=0.5,
    target=ax
)
plt.tight_layout()
```

#### 5.2.4.1. <a id='toc5_2_4_1_'></a>[节点层面指标（Node-level Metrics）](#toc0_)

```python
# 节点度 Degree
# 定义：某节点连接的边的数量，deg(v) = #edges connected to v
# 意义：衡量节点的直接连接能力或本地影响力
degree = g.degree()

# 介数中心性 Betweenness Centrality
# 定义：C_B(v) = Σ_{s≠v≠t} (σ_st(v) / σ_st)
# 意义：节点作为信息流的“中介”程度
betweenness = g.betweenness()

# 接近中心性 Closeness Centrality
# 定义：C_C(v) = 1 / Σ d(v, t)
# 意义：表示一个节点在网络中与其他节点的平均距离的倒数，中心程度越高，距离其他节点越近
closeness = g.closeness()

# 特征向量中心性 Eigenvector Centrality
# 定义：C_E(v) ∝ Σ_{u ∈ N(v)} C_E(u)
# 意义：不仅考虑连接数，还考虑连接对象的“重要性”，适合衡量“声望”或“传播力”
eigenvector = g.eigenvector_centrality()

# PageRank
# 定义：PageRank(v) = (1 - d)/N + d * Σ PageRank(u)/deg(u)  over neighbors u
# 意义：基于随机游走的中心性指标，Google最早用于网页排名
pagerank = g.pagerank()

# 局部聚类系数 Clustering Coefficient
# 定义：C(v) = 2 * (# of links between neighbors of v) / (deg(v) * (deg(v)-1))
# 意义：衡量邻居之间的连接程度（局部团结构）
clustering = g.transitivity_local_undirected(mode="zero")

# 打包为 DataFrame
df_nodes = pd.DataFrame({
    "Node": range(g.vcount()),
    "Degree": degree,
    "Betweenness": betweenness,
    "Closeness": closeness,
    "Eigenvector": eigenvector,
    "PageRank": pagerank,
    "ClusteringCoeff": clustering
})

df_nodes
```

#### 5.2.4.2. <a id='toc5_2_4_2_'></a>[边层面指标（Edge-level Metrics）](#toc0_)

```python
# 边介数中心性 Edge Betweenness
# 定义：C_B(e) = Σ_{s≠t} (σ_st(e) / σ_st)
# 意义：衡量某一边在连接不同网络区域时的“桥梁作用”
edge_betweenness = g.edge_betweenness()
g.es["betweenness"] = edge_betweenness

# 输出前5条边的指标值
for i, e in enumerate(g.es[:5]):
    print(f"Edge {e.tuple}: Betweenness = {e['betweenness']:.2f}")
```

## 5.3. <a id='toc5_3_'></a>[Screening the keystone taxa](#toc0_)

通常你已经有一个相关性矩阵（如SparCC, Spearman, Pearson），这里我举简单例子：

```python
import igraph as ig
import numpy as np


# 示例：10个微生物
adj_matrix = np.random.rand(10, 10)
adj_matrix = (adj_matrix + adj_matrix.T) / 2  # 对称化
adj_matrix[adj_matrix < 0.7] = 0              # 只保留强相关

g = ig.Graph.Weighted_Adjacency(adj_matrix.tolist(), mode="undirected", attr="weight", loops=False)
```

计算核心指标:

```python
# 计算各节点指标
degree = g.degree()
betweenness = g.betweenness()
closeness = g.closeness()
eigenvector = g.eigenvector_centrality()
```

赋值到节点属性，便于后面筛选：

```python
g.vs["degree"] = degree
g.vs["betweenness"] = betweenness
g.vs["closeness"] = closeness
g.vs["eigenvector"] = eigenvector
```

通常的标准可以是：
- 度数 > 平均度
- 介数中心性前 10%
- 特征向量中心性高
- 连到多个模块

core_microbes 是筛选后的节点子集。

```python
import numpy as np


threshold = np.mean(degree)

core_microbes = g.vs.select(degree_gt=threshold)
```

```python
import tensorflow as tf

w = tf.Variable(tf.constant(5, dtype=tf.float32)) # 初始值定为5

ETA = 0.1
EPOCH = 100

w_list = []
function = []
loss = []
for i in range(EPOCH):
    with tf.GradientTape() as tape:
        loss_function = tf.square(w + 1)
    grads = tape.gradient(loss_function, w) # 求导
    # print(type(grads))

    w.assign_sub(ETA*grads) # w = w - ETA*grads
    # w = tf.Variable(tf.constant(w - (ETA * grads), dtype=tf.float32)) # 同上

    print(f"w为：{w.numpy()}时, 函数值为：{loss_function}, 导数为：{grads}")
    
    # 接近w=-1时候的导数为0，函数值也为0
```
